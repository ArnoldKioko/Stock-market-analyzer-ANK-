```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('database/stocks.db')
print("Connected!")

# Query 1: Average closing price per stock
q1 = """
SELECT Ticker,
       ROUND(AVG(Close), 2) AS Avg_Close,
       ROUND(MAX(Close), 2) AS Max_Close,
       ROUND(MIN(Close), 2) AS Min_Close
FROM stock_prices
GROUP BY Ticker
ORDER BY Avg_Close DESC;
"""
print("📊 Average Prices:")
print(pd.read_sql(q1, conn))

# Query 2: Top 10 best daily return days
q2 = """
SELECT Date, Ticker, Close, Daily_Return_%
FROM stock_prices
ORDER BY Daily_Return_% DESC
LIMIT 10;
"""
print("\n🚀 Best Single-Day Returns:")
print(pd.read_sql(q2, conn))

# Query 3: Monthly average price per stock
q3 = """
SELECT Ticker,
       SUBSTR(Date, 1, 7) AS Month,
       ROUND(AVG(Close), 2) AS Avg_Monthly_Close
FROM stock_prices
GROUP BY Ticker, Month
ORDER BY Ticker, Month;
"""
print("\n📅 Monthly Averages:")
print(pd.read_sql(q3, conn))

conn.close()
```
